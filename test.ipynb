```python
import os

def install_python_3_10_12():
    # Step 1: Update system packages
    os.system("sudo apt update && sudo apt upgrade -y")

    # Step 2: Install required build dependencies
    os.system("sudo apt install -y software-properties-common")
    os.system("sudo apt install -y make build-essential libssl-dev zlib1g-dev "
              "libbz2-dev libreadline-dev libsqlite3-dev wget curl llvm "
              "libncurses5-dev libncursesw5-dev xz-utils tk-dev libffi-dev "
              "liblzma-dev python3-openssl git")

    # Step 3: Download Python 3.10.12 source
    os.system("cd /usr/src && sudo wget https://www.python.org/ftp/python/3.10.12/Python-3.10.12.tgz")

    # Step 4: Extract the tarball
    os.system("cd /usr/src && sudo tar xzf Python-3.10.12.tgz")

    # Step 5: Build and install
    os.system("cd /usr/src/Python-3.10.12 && sudo ./configure --enable-optimizations")
    os.system("cd /usr/src/Python-3.10.12 && sudo make altinstall")

    # Step 6: Set python3 and pip3 to point to the new version
    os.system("sudo update-alternatives --install /usr/bin/python3 python3 /usr/local/bin/python3.10 1")
    os.system("sudo update-alternatives --install /usr/bin/pip3 pip3 /usr/local/bin/pip3.10 1")
    
    # Step 7: Check the new versions
    os.system("python3 --version")
    os.system("pip3 --version")

if __name__ == "__main__":
    install_python_3_10_12()
```
